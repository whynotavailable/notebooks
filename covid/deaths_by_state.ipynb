# Some covid stats

```python
import pandas
import numpy as np
pandas.set_option('display.max_rows', None)
```

```python
df = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-23-2020.csv')
df['Confirmed'] = df['Confirmed'].astype(int)
df['Deaths'] = df['Deaths'].astype(int)
df['Recovered'] = df['Recovered'].astype(int)
print(df['Deaths'].sum())
df = df[df['Country_Region'] == 'US']
```

```python
def render_grouping(col):
    df2 = df.groupby(col).sum().sort_values(by=['Deaths'], ascending=False)
    df2["Row #"] = np.arange(len(df2)) + 1
    return df2.filter(items=['Row #', 'Confirmed', 'Deaths', 'Recovered'])
```

Here we have grouped by US state

```python
render_grouping('Province_State')
```

Grouping by us cities

```python
render_grouping('Combined_Key')[0:50]
```
